Specifically we will present the approach to:

 1) Calculate Simple Daily Intensity Index- Total monthly summed precipitation on wet days (days with  1mm), divided by total number of days in the month.
 2) Visualize the same index/indicator with the average values from 1990-2020 (extracted fromn ERA5) and 2021-2040 (extracted from the Climate DT)
 3) Visualize the standard deviation on monthly level concerning the averages from 1990-2020 and 2021-2040

## Data loading

```python
import xarray as xr
import pandas as pd
import os
from utils import precipitation
```

```python
data_prec = xr.open_dataset(
    "https://cacheb.dcms.destine.eu/d1-climate-dt/ScenarioMIP-SSP3-7.0-IFS-NEMO-0001-high-sfc-v0.zarr",
    storage_options={"client_kwargs":{"trust_env":"true"}},
    chunks={},
    engine="zarr",
)
```

```python
data_prec
```

```python
tp = data_prec.tp.astype("float32") *1000 #conversion from m to mm
tp.attrs["units"] = "mm"
```

```python
zagreb_lat_lon=(45.8130,15.9772)
dubrovnik_lat_lon=(42.6507, 18.0944)
```

Select your year of interest between 2020 and 2040.

```python
year_of_interest = "2035"
```

Decide whether you would like to perform a full analysis between 2020 and 2040.

```python
full_2020_2040_analysis = False # If True, the analysis is performed against the whole period, then comparing it to the year_of_interest
```

<div style="border: 2px solid #f5c518; background-color: #fff3cd; padding: 10px; border-radius: 5px;">
  <strong> Warning:</strong> The computation time for one city/pixel/point for 2020-2040 might require 1.5 hours or more. Therefore, <b>for initial experiments, we suggest keeping the variable full_2020_2040_analysis set to False.</b>
</div>

```python
# Select the data for Zagreb and Dubrovnik
tp_zagreb = tp.sel(latitude=zagreb_lat_lon[0], longitude=zagreb_lat_lon[1], method='nearest')
tp_dubrovnik = tp.sel(latitude=dubrovnik_lat_lon[0], longitude=dubrovnik_lat_lon[1], method='nearest')

if not full_2020_2040_analysis:
    # Subset the data for a given year
    tp_zagreb = tp_zagreb.sel(time=year_of_interest)
    tp_dubrovnik = tp_dubrovnik.sel(time=year_of_interest)
```

## SDII

```python
thresholds = [1] # select 1, 10 or 20

# Summarize the precipitation data for Zagreb and Dubrovnik
summary_zagreb, monthly_summary_zagreb = precipitation.summarize_precipitation(tp_zagreb, thresholds)
summary_dubrovnik, monthly_summary_dubrovnik = precipitation.summarize_precipitation(tp_dubrovnik, thresholds)
```

<div style="border: 2px solid #f5c518; background-color: #fff3cd; padding: 10px; border-radius: 5px;">
  <strong> Warning: </strong>The following cell usually takes <b>3 minutes</b> for each city/pixel/point. <b>If the variable full_2020_2040_analysis is set to True it might require 1.5 hour or more for one city/pixel/point.</b>
</div>

```python
# Convert the monthly summaries to pandas DataFrames for easier viewing
monthly_summary_zagreb_df = monthly_summary_zagreb.to_dataframe()
monthly_summary_dubrovnik_df = monthly_summary_dubrovnik.to_dataframe()
```

```python
# Reset the index to make 'valid_time' a column
monthly_summary_zagreb_df = monthly_summary_zagreb_df.reset_index()
monthly_summary_dubrovnik_df = monthly_summary_dubrovnik_df.reset_index()

# Calculate the number of days in each month
def days_in_month(month):
    return month.days_in_month

# Calculate SDII for Zagreb
monthly_summary_zagreb_df['days_in_month'] = monthly_summary_zagreb_df['time'].dt.to_period('M').apply(days_in_month)
sdii_zagreb = monthly_summary_zagreb_df['monthly_sums'] / monthly_summary_zagreb_df['days_in_month']
sdii_zagreb_df = pd.DataFrame({'month': monthly_summary_zagreb_df['time'].dt.to_period('M'), 'SDII': sdii_zagreb})
    
# Calculate SDII for Dubrovnik
monthly_summary_dubrovnik_df['days_in_month'] = monthly_summary_dubrovnik_df['time'].dt.to_period('M').apply(days_in_month)
sdii_dubrovnik = monthly_summary_dubrovnik_df['monthly_sums'] / monthly_summary_dubrovnik_df['days_in_month']
sdii_dubrovnik_df = pd.DataFrame({'month': monthly_summary_dubrovnik_df['time'].dt.to_period('M'), 'SDII': sdii_dubrovnik})
```

```python
# Save the results - this is useful later in case you would like to skip to the computation step above and play with the data
csv_directory = "csv"
if not os.path.exists(csv_directory):
    os.makedirs(csv_directory)
    
sdii_zagreb_df.to_csv(f'{csv_directory}/sdii_zagreb.csv', sep=',', index=False)
sdii_dubrovnik_df.to_csv(f'{csv_directory}/sdii_dubrovnik.csv', sep=',', index=False)
```

Read back the CSVs and use ERA5 data as a reference:

```python
sdii_zagreb_df = pd.read_csv(f'{csv_directory}/sdii_zagreb.csv', parse_dates=['month'])
sdii_dubrovnik_df = pd.read_csv(f'{csv_directory}/sdii_dubrovnik.csv', parse_dates=['month'])

reference_directory = "reference"
sdii_zagreb_df_era5 = pd.read_csv(f'{reference_directory}/sdii_zagreb_ERA5.csv', parse_dates=['month'])
sdii_dubrovnik_df_era5 = pd.read_csv(f'{reference_directory}/sdii_dubrovnik_ERA5.csv', parse_dates=['month'])

# Set 'month' as the index
sdii_zagreb_df.set_index('month', inplace=True)
sdii_zagreb_df_era5.set_index('month', inplace=True)
sdii_dubrovnik_df.set_index('month', inplace=True)
sdii_dubrovnik_df_era5.set_index('month', inplace=True)
```

```python
# Extract data for the year of interest
data_selected_zagreb = sdii_zagreb_df[sdii_zagreb_df.index.year == int(year_of_interest)]
data_selected_dubrovnik = sdii_dubrovnik_df[sdii_dubrovnik_df.index.year == int(year_of_interest)]


# Calculate the average and std for the whole period and for the ERA5 data
sdii_zagreb_df['month_num'] = sdii_zagreb_df.index.month
sdii_zagreb_df_era5['month_num'] = sdii_zagreb_df_era5.index.month

avg_sdii_zagreb = sdii_zagreb_df.groupby('month_num')['SDII'].mean()
std_sdii_zagreb = sdii_zagreb_df.groupby('month_num')['SDII'].std()

avg_sdii_zagreb_era5 = sdii_zagreb_df_era5.groupby('month_num')['SDII'].mean()
std_sdii_zagreb_era5 = sdii_zagreb_df_era5.groupby('month_num')['SDII'].std()

sdii_dubrovnik_df['month_num'] = sdii_dubrovnik_df.index.month
sdii_dubrovnik_df_era5['month_num'] = sdii_dubrovnik_df_era5.index.month

avg_sdii_dubrovnik = sdii_dubrovnik_df.groupby('month_num')['SDII'].mean()
std_sdii_dubrovnik = sdii_dubrovnik_df.groupby('month_num')['SDII'].std()

avg_sdii_dubrovnik_era5 = sdii_dubrovnik_df_era5.groupby('month_num')['SDII'].mean()
std_sdii_dubrovnik_era5 = sdii_dubrovnik_df_era5.groupby('month_num')['SDII'].std()
   
# Plot the data for Zagreb -SDII
precipitation.plot_data(
    city_name="Zagreb",
    data_selected=data_selected_zagreb,
    avg_data=avg_sdii_zagreb,
    std_data=std_sdii_zagreb,
    avg_era5=avg_sdii_zagreb_era5,
    std_era5=std_sdii_zagreb_era5,
    yoi=year_of_interest,
    ylabel="Simple Daily Intensity Index",
    column_name="SDII"
)

# Plot the data for Dubrovnik - SDII
precipitation.plot_data(
    city_name="Dubrovnik",
    data_selected=data_selected_dubrovnik,
    avg_data=avg_sdii_dubrovnik,
    std_data=std_sdii_dubrovnik,
    avg_era5=avg_sdii_dubrovnik_era5,
    std_era5=std_sdii_dubrovnik_era5,
    yoi=year_of_interest,
    ylabel="Simple Daily Intensity Index",
    column_name="SDII"
)
```
